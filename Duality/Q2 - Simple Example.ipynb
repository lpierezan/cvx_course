```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
import matplotlib

%matplotlib inline
```

$L(x,\lambda) = x^2 + 1 + \lambda(x^2 -6x + 8)$
$= (\lambda+1)x^2 -6x\lambda + 1 + 8\lambda$  

$x_{min} = \frac{6\lambda}{2(\lambda+1)} = \frac{3\lambda}{\lambda + 1}$

$g(\lambda) = \inf_{x \in D} L(x,\lambda) = L(x_{min}, \lambda) = $
$\frac{-9\lambda^2}{\lambda+1} + 8\lambda + 1$


```python
def f0(x):
    return x**2 + 1
def f1(x):
    return (x-2)*(x-4)
def L(x,l):
    return f0(x) + l*f1(x) 

def g(l):
    xMin = 3*l/(l+1)
    return xMin , L(xMin, l)

x = np.linspace(-1,5,100)
obj = f0(x)
fig, ax = plt.subplots(figsize = (15,10))
ax.plot(x,obj, label = 'objective')
for i, l in enumerate([0.1,0.5,1,2,4,8]):
    ax.plot(x, L(x,l), label = f'lambda {l}', c = f"C{i+1}")
    xMin,lMin = g(l)
    ax.scatter(xMin, lMin, s=80, marker = '*', c = f"C{i+1}")

ax.vlines([2,4], 0, max(obj), linestyles='dashed', label = 'feasible')
ax.set_ylim(0,max(obj))
ax.legend()

plt.show()
```

```python
# strong duality
g(2)[1] == f0(2)
```
